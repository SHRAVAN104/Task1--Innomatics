## **Q1. User Login Check**

```python
def login_check(username, password):
    # Stored credentials
    correct_username = "admin"
    correct_password = "1234"
    
    if username == correct_username and password == correct_password:
        return "Login Successful"
    else:
        return "Invalid Credentials"


# Test credentials
username = "admin"
password = "1234"

result = login_check(username, password)
print(result)
```

## **Q2. Pass / Fail Analyzer**

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total   Number Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

## **Q3. Simple Data Cleaner**

```python

names = [" Alice ", "bob", " CHARLIE "]

# Clean names: remove spaces and convert to lowercase
cleaned_names = [name.strip().lower() for name in names]

print("Cleaned Names:", cleaned_names)
```

## **Q4. Message Length Analyzer**

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    if length > 10:
        print("→ This message is longer than 10 characters")
```

## **Q5. Error Message Detector**

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count ERROR entries
error_count = logs.count("ERROR")

print("Total ERROR messages:", error_count)
```
